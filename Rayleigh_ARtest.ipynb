```python
%matplotlib inline
import matplotlib.pyplot as plt
from pytmatrix.tmatrix import Scatterer
from pytmatrix.psd import PSDIntegrator, GammaPSD, BinnedPSD
from pytmatrix import orientation, radar, tmatrix_aux, refractive
import numpy as np
import scipy.io as sio 
```

```python
nconc = 100
ars = np.linspace(0.2,1, num = 100)
dbase = 1
flec1, flec2 = [], []

#Z = (.176/.93*(6/(np.pi)/.917)**2*nconc*(.004*(dbase*0.1)**2.1)**2)*10**6;
#Rayleigh = 10*np.log10(Z)

for ar in ars:
    dmod = dbase * (ar**(1./3.))
    oblate = Scatterer(radius = dmod/2.,wavelength=tmatrix_aux.wl_C, m=refractive.m_w_10C[tmatrix_aux.wl_C],  axis_ratio = 1/ar)
    flec1.append(10*np.log10(nconc*radar.refl(oblate)))
    Z2 = oblate.get_SZ_single()[1]
    hZ2 = 2 * np.pi * (Z2[0,0] - Z2[0,1] - Z2[1,0] + Z2[1,1])
    flective = oblate.wavelength**4/(np.pi**5*oblate.Kw_sqr) *hZ2
    flec2.append(10*np.log10(nconc*flective))
#Z = (.176/.93*(6/(np.pi)/.917)**2*nconc*(.004*(dbase*0.1)**2.1)**2)*10**6;
Z = (dbase**6)
Rayleigh = 10*np.log10(Z*nconc)    

plt.plot(ars, np.array(flec1) - Rayleigh*np.ones(np.shape(ars)))
plt.xlabel('Axial Ratio')
plt.ylabel('Reflectivity (Db)')
plt.title('Deviation from Rayleigh reflectivity for 1000 particles')
plt.legend(['Orientation averaged', 'Rayleigh Scattering'], loc = 0)
plt.show()

plt.plot(ars, np.array(flec2) - Rayleigh*np.ones(np.shape(ars)))
plt.xlabel('Axial Ratio')
plt.ylabel('Reflectivity (Db)')
plt.title('Deviation from Rayleigh reflectivity for 1000 particles')
plt.legend(['Single Orientation', 'Rayleigh Scattering'], loc = 0)
plt.show()
```
