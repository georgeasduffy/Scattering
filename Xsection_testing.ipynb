```python
%matplotlib inline
import matplotlib.pyplot as plt
from pytmatrix.tmatrix import Scatterer
from pytmatrix.psd import PSDIntegrator, GammaPSD, BinnedPSD
from pytmatrix import orientation, radar, tmatrix_aux, refractive
import numpy as np
import scipy.io as sio 
import pandas
import pdb
```

```python
def mix(a, b, d, ar):
    # a, d in cgs, use bruggeman equation
    m = a*d**b
    v = (1./6.)*ar*(d)**3*np.pi
    density = m/v
    n = 1.782
    kappa = complex(0, 7.302*10**-3)
    IceDens = .917
    e1 = n**2 - kappa**2
    e2 = 2*n*kappa
    B = e1 + e2
    f = density/IceDens
    mixed = (B*(1+2*f)-(2*f-2))/((2+f)+B*(1-f))
    e1out = np.real(mixed)
    e2out = np.imag(mixed)
    nout = np.sqrt((np.sqrt(e1out**2 +e2out**2)+e1out)/2)
    kout = np.sqrt((np.sqrt(e1out**2 +e2out**2)-e1out)/2)
    m =complex(nout, kout)
    return m
```

```python
nconc = 100
ars = np.linspace(0.2,1, num = 100)
bees = np.linspace(1.8, 2.4, num = 100)
ays = np.linspace(0.002, 0.012, num = 100) 
dees = np.logspace(-1, 1, num = 100)
dbase = 1
flec1, flec2, flec3, flec4 = [], [], [], []

Z = (.176/.93*(6/(np.pi)/.917)**2*nconc*(.004*(dbase*0.1)**2.1)**2)*10**6;
Rayleigh = 10*np.log10(Z)
```

```python
for ar in ars:
    dmod = dbase * (ar**(1./3.))
    mm = mix(0.004, 2.1, dbase*0.1, ar)
    oblate = Scatterer(radius = dmod/2., wavelength = 1000., m = mm, axis_ratio = 1/ar)
    flec1.append(10*np.log10(nconc*radar.refl(oblate)))
    
Z = (.176/.93*(6/(np.pi)/.917)**2*nconc*(.004*(dbase*0.1)**2.1)**2)*10**6;
Rayleigh = 10*np.log10(Z)    

plt.plot(ars, np.array(flec1) - Rayleigh*np.ones(np.shape(ars)))
plt.xlabel('Axial Ratio')
plt.ylabel('Reflectivity (Db)')
plt.title('Deviation from Rayleigh reflectivity for 1000 particles')
plt.legend(['T-matrix simulations', 'Rayleigh Scattering'], loc = 0)
plt.show()

```

```python
Rayleigh = []
for d in dees:
    ar = 0.6
    dmod = d * (ar**(1./3.))
    mm = mix(0.004, 2.1, d*0.1, ar)
    oblate = Scatterer(radius = dmod/2., wavelength = 1000., m = mm, axis_ratio = 1/ar)
    flec2.append(10*np.log10(nconc*radar.refl(oblate)))
    Z = (.176/.93*(6/(np.pi)/.917)**2*nconc*(.004*(d*0.1)**2.1)**2)*10**6;
    Rayleigh.append(10*np.log10(Z))

plt.plot(dees, np.array(flec2) - np.array(Rayleigh))
plt.xlabel('diameter (mm)')
plt.ylabel('reflectivity (Db)')
plt.title('Deviation from Rayleigh reflectivity for 1000 particles')
plt.show()
```
